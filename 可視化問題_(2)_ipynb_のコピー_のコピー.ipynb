## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

問1.png

```python
#CSVの読み込み
dataset = pd.read_csv("drive/My Drive/Colab Notebooks/matplotlib_data/hist_data.csv")

#listに変換
data_a = dataset["aデータ"].to_list()
data_b = dataset["bデータ"].to_list()

#ヒストグラムの作成
plt.title("aデータとbデータの温度に関する度数分布")  #タイトル追加
plt.xlabel("温度(℃)", fontsize=10)  #x座標のラベル追加
plt.ylabel("度数", fontsize=10)  #y座標のラベル追加
plt.hist(data_a, bins=30, alpha=0.4, color="darkorange", label="aデータ")  #aデータのグラフ作成
plt.hist(data_b, bins=30, alpha=0.4, color="c", label="bデータ")  #bデータのグラフ作成
plt.legend(loc="upper right", fontsize=10)  #凡例追加

#ヒストグラムの表示
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
#CSVの読み込み
dataset = pd.read_csv("drive/My Drive/Colab Notebooks/matplotlib_data/scatter_data.csv")


#データの整理
class0 = dataset.iloc[2:50, :]  #classが0のデータを取得(query使えない？)
class1 = dataset.iloc[50:100, :]
class2 = dataset.iloc[100: , :]

dataX_class0 = class0["length (cm)"].to_list()  #class0のX座標
dataY_class0 = class0["width (cm)"].to_list()  #class0のY座標

dataX_class1 = class1["length (cm)"].to_list()
dataY_class1 = class1["width (cm)"].to_list()

dataX_class2 = class2["length (cm)"].to_list()
dataY_class2 = class2["width (cm)"].to_list()

#散布図を作成
plt.title("クラス0〜2のlengthとwidthに関する散布図")  #タイトル追加
plt.xlabel("length (cm)", fontsize=10)  #X座標のラベル追加
plt.ylabel("width (cm)", fontsize=10)  #Y座標のラベル追加
plt.grid(True)  #目盛線表示
plt.scatter(dataX_class0, dataY_class0, label="class0", color="r")  #class0のグラフを表示
plt.scatter(dataX_class1, dataY_class1, label="class1", color="b")
plt.scatter(dataX_class2, dataY_class2, label="class2", color="g")
plt.legend(loc="upper right", fontsize=10)  #凡例追加

#グラフを表示
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

問3(1).png

問3(2).png

```python
#CSVの読み込み
dataset = pd.read_csv("drive/My Drive/Colab Notebooks/matplotlib_data/line_data.csv")

#listに変換
data_X = dataset["期間"].to_list()
data_Y = dataset["件数"].to_list()

#グラフの作成
plt.title("期間ごとの件数の推移")  #タイトル追加
plt.xlabel("期間", fontsize=10)  #X座標のラベル追加
plt.ylabel("件数", fontsize=10)  #Y座標のラベル追加
plt.plot(data_X, data_Y, marker="o")  #折れ線グラフを作成
plt.xticks(rotation=30)  #X座標のラベルを30回転
plt.grid(True)  #目盛線表示

#グラフの表示
plt.show()






#listに変換
data_Y2 = dataset["費用"].to_list()

#棒グラフの作成
plt.title("期間ごとの費用の推移")  #タイトル追加
plt.xlabel("期間", fontsize=10)  #X座標のラベル追加
plt.ylabel("費用(円)", fontsize=10)  #Y座標のラベル追加
plt.bar(data_X, data_Y2)
plt.grid(True)  #目盛線表示
plt.xticks(rotation=30)  #X座標のラベルを30回転

#グラフの表示
plt.show()
```
